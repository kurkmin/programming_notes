# Object-Oriented Programming (OOP)

def Object-Oriented Programming (OOP): writing code directed toward modelling objects
def Object: a collection of data and associated behaviours 

There are three stages in object-oriented software development: 

1. def Object-Oriented Analysis (OOA): the process of looking at a problem, system or task and identifying the objects and interactions between those objects;At this stage, the output of analysis is in the form of requirements. 

2. def Object-Oriented Design (OOD): the prcoess of converting such requirements into an implementation specification; the designer should name the objects, define the behaviours, and formally specify which objects can activate specific behaviours on other objects. The output of the design stage is an implementation specification.
